## Optimize the Model


Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Check out the NAME value counts for relevancy and for binning
name_count = application_df['NAME'].value_counts()
name_count
```

```python
# Check out The EIN value counts for relevancy and for binning
ein_count = application_df["EIN"].value_counts()
ein_count
```

```python
# Drop the non-beneficial ID column 'EIN', and keep the NAME column.
new_app_df = application_df.drop(columns=['EIN'])
new_app_df.head()
```

```python
# Determine the number of unique values in each column.
new_app_df.nunique()
```

```python
# Choose a cutoff value and create a list of name to be replaced
# use the variable name `name_to_replace`
name_to_replace = list(name_count[name_count <10].index)

# Replace in dataframe
for app in name_to_replace:
    new_app_df['NAME'] = new_app_df['NAME'].replace(app,"Other")

# Check to make sure binning was successful
new_app_df['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_count = new_app_df['APPLICATION_TYPE'].value_counts()
app_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_count[app_count <500].index)

# Replace in dataframe
for app in application_types_to_replace:
    new_app_df['APPLICATION_TYPE'] = new_app_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
new_app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classif_count = new_app_df['CLASSIFICATION'].value_counts()
classif_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classif_count[classif_count > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(classif_count[classif_count< 500].index)

# Replace in dataframe
for cls in classifications_to_replace:
    new_app_df['CLASSIFICATION'] = new_app_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
new_app_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df_clean = pd.get_dummies(new_app_df)
df_clean.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = df_clean['IS_SUCCESSFUL'].values
X = df_clean.drop(columns=['IS_SUCCESSFUL']).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Check the number of colunms for X_train to be fed as input
len(X_train[0])
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =100
hidden_nodes_layer2 = 50
hidden_nodes_layer3 = 30
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation="relu", input_dim=number_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="tanh"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_nn= nn.fit(X_train_scaled, y_train, epochs=200)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```
